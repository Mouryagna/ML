```python
import pandas as pd
import requests
import time
```

```python
url='https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page=1'
response=requests.get(url)
```

```python
df=pd.DataFrame(response.json()['results'])
df.head()
```

```python
df1=df[['id','title','overview','release_date','original_language','vote_average','vote_count']]
df1.head()
```

```python
df = pd.DataFrame()
df
```

```python
for i in range(1,429):
    try:
        url='https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={}'.format(i)
        response=requests.get(url)
        temp_df=pd.DataFrame(response.json()['results'])[['id','title','overview','release_date','original_language','vote_average','vote_count']]
        df = pd.concat([df, temp_df], ignore_index=True)
        time.sleep(0.3)
    except:
        print(f"skipped {i}")
df
```
