```python
#use cols paramter
#we can filter the cols that what we need
pd.read_csv('csv/data.csv',usecols=['date','price','city'])
```

```python
#squeeze paramter converts that into series
pd.read_csv('csv/data.csv',usecols=['date'],squeeze=True)
```

```python
#skiprows /nrows will skip rows/ give that match size of rows
pd.read_csv('csv/data.csv',skiprows=[1,2])
pd.read_csv('csv/data.csv',nrows=100)
```

```python
#Encoding like to save data to get data from it can know which type of encode ex
pd.read_csv('csv/zomato.csv',encoding='Latin-1')
```

```python
#Skip bad lines for ex proper cols 4 and some rows get 5 or more cols we can skip
pd.read_csv('csv/data.csv',error_bad_lines=False)
```

```python
#handling dates
pd.read_csv('csv/data.csv',parse_dates=['date']).info()
```

```python
#convertors
def rename(name):
    if name=='Shoreline':
        return 'SH'
    else:
        return name
```

```python
pd.read_csv('csv/data.csv',converters={'city':rename})
```

```python
#na_values paramter which we can make nan values in dataset
pd.read_csv('csv/data.csv',na_values=['Shoreline']).head()
```

```python
#loading a huge daataset iiin chunks
dfs=pd.read_csv('csv/data.csv',chunksize=1000)
for chunks in dfs:
    print(chunks.shape)
```
