```python
extract_8_15=df['feature_8,feature_15'].str.split(',',expand=True)
df['feature_8']=extract_8_15[0]
df['feature_15']=extract_8_15[1]

extract_21_10=df['feature_21,feature_10'].str.split(',',expand=True)
df['feature_21']=extract_21_10[0]
df['feature_10']=extract_21_10[1]

extract_1_6=df['feature_1,feature_6'].str.split(',',expand=True)
df['feature_1']=extract_1_6[0]
df['feature_6']=extract_1_6[1]
```

```python
for col in df.columns:
    df[col] = df[col].str.replace(r'[^a-zA-Z]', '', regex=True)
```

```python
df.drop(columns=['feature_8,feature_15','feature_21,feature_10','feature_1,feature_6'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split

x=df.drop(columns=['class'])
y=df['class']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
trf=[trf1,trf2,trf3,trf4,trf5,trf6,trf7,trf8,trf9,trf10,trf11,trf12,trf13,trf14,trf15,trf16,trf17,trf18,trf19,trf20,trf21,trf22]
```

```python
transformer=ColumnTransformer(transformers=trf,remainder='passthrough')
```

```python
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report

lr=LogisticRegression(random_state=42)
pipe=Pipeline([
    ('transformer',transformer),
    ('clf',lr)
])
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
accuracy_score(y_pred,y_test)
```
