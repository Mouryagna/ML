# EDA using Bivariate and Multivariate

```python
import pandas as pd
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
titanic=pd.read_csv('csv/train.csv')
flights=sns.load_dataset('flights')
iris=sns.load_dataset('iris')
```

```python
tips.sample(5)
```

## 1.ScatterPlot (N-N)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'],style=tips['day'],size=tips['size'])
```

## 2.Barplot (N-C)

```python
titanic.sample(5)
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Survived'],hue=titanic['Sex'],palette=['blue','pink'])
```

## 3.Boxplot (N-C)

```python
titanic.sample(5)
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

## 4.Distplot (N-C)

```python
titanic.head()
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

## 5.Heatmap (C-C)

```python
sns.heatmap(pd.crosstab(titanic['Survived'],titanic['Pclass']))
```

```python
titanic.groupby('Pclass').sum()['Survived'].plot(kind='bar')
```

## 6.ClusterMap (C-C)

```python
sns.clustermap(pd.crosstab(titanic['SibSp'],titanic['Survived']))
```

## 7.Pairplot

```python
sns.pairplot(iris,hue='species')
```

## 8.LinePlot (N-N)

```python
flights.head()
```

```python
new=flights.groupby('year').sum(numeric_only=True).reset_index()
new
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
